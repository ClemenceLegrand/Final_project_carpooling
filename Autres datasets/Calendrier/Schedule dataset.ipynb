```python
import pandas as pd
import os
```

```python
row_data = '/Users/clemence/IRONHACK/Final Project/Autres datasets/Calendrier'
all_files = [] 
tracking_files=[]

for filename in os.listdir(row_data):
    if filename.endswith('.csv'):
        filepath = os.path.join(row_data, filename)
        tracking_files.append(filename)
        file = pd.read_csv(filepath,delimiter=';')
        all_files.append(file)

schedule_df = pd.concat(all_files, axis=0, ignore_index=True)
```

```python
schedule_df
```

```python
schedule_df.dtypes
```

```python
schedule_df["Date"]=pd.to_datetime(schedule_df["Date"])
```

```python
schedule_df["Weekday"]=schedule_df['Date'].dt.day_name()
```

```python
schedule_df.head()
```

```python
schedule_df=schedule_df.sort_values(by=["Date"])
```

```python
schedule_df.head(20)
```

```python
schedule_df["Calendrier Zone A"].value_counts()
```

```python
schedule_df = schedule_df.drop(["timestamp_unix"], axis=1)
```

```python
schedule_df = schedule_df.rename(columns={"Date": "id_date"})
```

```python
schedule_df[['Calendrier Zone A','Calendrier Zone B','Calendrier Zone C']] = schedule_df[['Calendrier Zone A','Calendrier Zone B','Calendrier Zone C']].replace(["Hors Vacances","Vacances de Noël", "Vacances d'été","Vacances d'hiver","Vacances de printemps","Vacances de la Toussaint","Pont de l'Ascension"], [0, 1, 1, 1, 1,1,1])
```

```python
schedule_df.head(20)
```

```python
schedule_df["vacation_level"]=schedule_df["Calendrier Zone A"]+schedule_df["Calendrier Zone B"]+schedule_df["Calendrier Zone C"]
```

```python
schedule_df.head(20)
```

```python
schedule_df = schedule_df.drop(["Calendrier Zone A","Calendrier Zone B","Calendrier Zone C"], axis=1)
```

```python
schedule_df.head(20)
```

```python
schedule_df.rename(columns={"id_date":"date""Nom de la période":"holidays_name","Weekday": "weekday""},inplace=True)
```

```python
schedule_df["Nom de la période"].fillna("Hors vacances",inplace=True)
```
