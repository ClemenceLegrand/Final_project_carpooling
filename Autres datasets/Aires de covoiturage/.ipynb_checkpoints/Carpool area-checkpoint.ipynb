```python
counts=pd.Series(park_df.isnull().sum())
```

```python
columns_to_drop=counts[counts>200].index.tolist()
```

```python
columns_to_drop
```

```python
column_indices = [park_df.columns.get_loc(col) for col in columns_to_drop]

park_df = park_df.drop(park_df.columns[column_indices], axis=1)
```

```python
park_df.head(10)
```

```python
park_df["ouvert"].value_counts()
```

```python
park_df = park_df.drop("ouvert", axis=1)
```

```python
park_df = park_df.drop("source", axis=1)
```

```python
park_df.head(10)
```

```python
park_df.rename(columns={"id_lieu":"id","nom_lieu": "name","com_lieu":"city_name","insee":"id_city","date_maj":"date","Xlong":"long","Ylat":"lat"},inplace=True)
```
